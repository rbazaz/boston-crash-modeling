```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd
df=pd.read_csv("testing.csv",index_col=0)
```

```python
places=df['Place'].unique()
```

```python
vehicles = ['2 Axle 6 Tire', u'2 Axle Long', u'3 Axle Single', u'4 Axle Single',
       u'5 Axle Double', u'6 Axle Multi', u'<5 Axle Double', u'<6 Axle Multi',
       u'>6 Axle Double', u'>6 Axle Multi', u'Buses', u'Cars & Trailers','Motor Cycles']
```

```python
df.groupby("Place").sum()
```

```python
distribution_place=[]
for i in places:
    temp_df=df[df.Place==i]
    distribution_place+=[(i,sum([sum(row) for i, row in temp_df[vehicles].iterrows()]))]
```

```python
df=pd.DataFrame(distribution_place)
```

```python
df.columns = [['Places',"Distribution"]]
df
```

```python
ax=sns.barplot(y=df.Places, x = df.Distribution)
ax.set_title("Distribution across Place",fontsize=30)
```
